La neumonía es una infección aguda respiratoria que afecta los pulmones. Es la principal causa individual de mortalidad infantil a nivel mundial, responsable del 15% de todas las defunciones en menores de 5 años, según la Organización Mundial de la Salud (OMS). Más del 90% de los casos de neumonía clínica se diagnostican en países en desarrollo, donde los recursos médicos son limitados. El diagnóstico de esta enfermedad requiere el uso de radiografías de tórax, cuyo análisis debe ser realizado por médicos altamente capacitados y experimentados, lo que dificulta la detección temprana y el tratamiento.

Un problema recurrente en la detección de enfermedades pulmonares es la similitud de los patrones característicos con otras enfermedades, lo que dificulta un diagnóstico preciso por parte de los médicos especializados.

Entre las técnicas de deep learning, las redes neuronales convolucionales han demostrado ser eficaces en la clasificación de imágenes y son frecuentemente utilizadas por los investigadores en visión computacional. Estas técnicas han demostrado ser tan útiles para el reconocimiento rápido y preciso de la neumonía que, en algunos casos, igualan o superan a un especialista médico. Sin embargo, es importante destacar que estas técnicas no pueden reemplazar a los médicos especializados en el diagnóstico, sino que son herramientas complementarias para el personal médico en la toma de decisiones clínicas (Hosny et al., 2018; Kallianos et al., 2019).


La inteligencia artificial (IA) posee un potencial revolucionario en el ámbito del diagnóstico y tratamiento de enfermedades, especialmente en la interpretación de imágenes médicas complejas. Aunque prometedor, la aplicabilidad clínica y la implementación práctica de la IA aún presentan desafíos significativos para los profesionales de la salud.

Tradicionalmente, el análisis de imágenes médicas se basaba en un enfoque algorítmico que implicaba varios pasos laboriosos, desde la segmentación manual de objetos hasta la clasificación utilizando clasificadores estadísticos o de aprendizaje automático. Sin embargo, este método era intensivo en recursos humanos y computacionales.

El desarrollo de las redes neuronales convolucionales ha marcado un avance crucial en la capacidad de la IA para clasificar imágenes y detectar objetos en ellas. Estas redes aplican filtros de análisis de imágenes en múltiples capas de procesamiento, permitiendo una representación abstracta de las imágenes que facilita su clasificación.

Una técnica importante para abordar la escasez de datos en ciertos dominios es el aprendizaje por transferencia, que consiste en aprovechar datos de dominios similares para entrenar modelos de IA. Esta técnica ha demostrado ser particularmente efectiva en situaciones donde los datos son limitados.

Aplicando este enfoque al diagnóstico de la neumonía pediátrica, una enfermedad que afecta gravemente a los niños en países en desarrollo, se busca mejorar la precisión y rapidez del diagnóstico. La neumonía, que causa un gran número de muertes infantiles cada año, puede ser de origen bacteriano o viral, requiriendo enfoques de tratamiento diferentes.

El uso de imágenes de radiografías de tórax pediátricas para la detección de neumonía y la diferenciación entre sus tipos puede ser fundamental en entornos donde los recursos son limitados y se necesita una atención médica urgente. En este estudio, se recolectaron y etiquetaron miles de imágenes para entrenar el sistema de IA, y los resultados preliminares muestran una prometedora capacidad para detectar esta enfermedad.

El conjunto de datos se organiza en tres carpetas principales: "train" (entrenamiento), "test" (prueba) y "val" (validación), cada una de las cuales contiene subcarpetas para las dos categorías de imágenes disponibles: "Neumonía" y "Normal". En total, hay 5,863 imágenes de rayos X en formato JPEG, divididas en las categorías de neumonía y normal.

Estas imágenes de radiografías de tórax se seleccionaron de cohortes retrospectivas de pacientes pediátricos, con edades comprendidas entre uno y cinco años, del Centro Médico de Mujeres y Niños de Guangzhou, Guangzhou. Todas estas radiografías de tórax se realizaron como parte de la atención clínica de rutina de los pacientes.

Para garantizar la calidad de las imágenes utilizadas en el análisis, se llevó a cabo un proceso de control de calidad inicial, donde se eliminaron las exploraciones de baja calidad o ilegibles. Además, los diagnósticos asociados a las imágenes fueron evaluados por dos médicos expertos antes de ser autorizados para el entrenamiento del sistema de inteligencia artificial (IA). Para mitigar posibles errores de evaluación, un tercer experto revisó el conjunto de datos de evaluación para asegurar la precisión de los diagnósticos.

En el presente trabajo, se exploran diferentes arquitecturas de redes neuronales convolucionales (CNN), denominadas model1, model2 y model3, para clasificar imágenes de radiografías de tórax en dos clases: normal y neumonía. Para estos tres modelos entrenados, se importarán los pesos en archivos HDF5. Luego, se definirán los generadores de datos para el entrenamiento y la validación utilizando ImageDataGenerator. Posteriormente, se empleará Keras Tuner para buscar los hiperparámetros óptimos y crear un modelo ensamblado que combine los tres modelos entrenados.

Se evaluará el rendimiento del modelo ensamblado utilizando el conjunto de prueba, generando métricas de evaluación, informes de clasificación y matrices de confusión. Los modelos se evaluarán individualmente, así como el modelo ensamblado, utilizando métricas de rendimiento, y se compararán los resultados obtenidos.

Las predicciones del modelo ensamblado se guardarán en un archivo pickle para su uso futuro, y los modelos ensamblados se guardarán en archivos HDF5 y Keras para su despliegue y uso en entornos de producción.

Se preparan los datos de imágenes de radiografías de tórax para su uso en el entrenamiento y evaluación de un modelo de aprendizaje automático.
```python
# Obtenemos la ruta de acceso a los conjuntos de datos


# Asignar las rutas proporcionadas a las variables
train_path = "C:/Users/marce/Documents/Bootcamp_UDD/DS_AI/MODULO 7/Proyecto_M7/data/chest_xray/train"
val_path = "C:/Users/marce/Documents/Bootcamp_UDD/DS_AI/MODULO 7/Proyecto_M7/data/chest_xray/val"
test_path = "C:/Users/marce/Documents/Bootcamp_UDD/DS_AI/MODULO 7/Proyecto_M7/data/chest_xray/test"


# Listar archivos y directorios en las rutas de los conjuntos de datos
print("Entrenamiento:", os.listdir(train_path))
print("Validación:", os.listdir(val_path))
print("Prueba:", os.listdir(test_path))
```

```python
# Definir el número de épocas para entrenar el modelo 1, modelo 2 y modelo 3
epochs = 10

# significa que cada lote contendrá 32 imágenes
batch_size = 32
```

Un **ImageDataGenerator** es una clase de Keras que se utiliza para la generación de lotes de datos de imágenes de tensor. Esta herramienta es especialmente útil en el dominio del aumento de datos en tiempo real, conocido como *data augmentation*.

Los beneficios de usar un ImageDataGenerator incluyen:

- **Generación de lotes**: Permite procesar imágenes en lotes más pequeños, lo cual es útil cuando se trabaja con conjuntos de datos grandes o imágenes de alta resolución que podrían no caber en la memoria si se procesan todas a la vez.
- **Aumento de datos**: Puede realizar transformaciones en las imágenes como rotación, zoom, volteo horizontal, entre otros, para aumentar el tamaño del conjunto de datos y mejorar la generalización del modelo.
- **Preprocesamiento**: Ofrece capacidades de preprocesamiento de imágenes, lo que facilita la normalización y la aplicación de filtros antes de alimentar las imágenes a la red neuronal.

```python
# Definir generadores de datos para entrenamiento, validación y prueba

train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_path,
    target_size=(150, 150),
    batch_size=batch_size,
    class_mode='binary'  # O 'categorical' si hay más de dos clases
)

val_generator = val_datagen.flow_from_directory(
    val_path,
    target_size=(150, 150),
    batch_size=batch_size,
    class_mode='binary'
)

test_generator = test_datagen.flow_from_directory(
    test_path,
    target_size=(150, 150),
    batch_size=batch_size,
    class_mode='binary'
)
```

Los modelos se entrenan utilizando datos de entrenamiento y se validan utilizando datos de validación para evaluar su rendimiento durante el entrenamiento. ### Construir un modelo convolucional con MODELO 1

```python
model1 = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')  # Capa de salida para clasificación binaria
])
```

```python
model1.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

### Comprobar la arquitectura de capas MODELO 1

```python
# Resumen del modelo
model1.summary()
```

### Entrenamiento del MODELO 1

```python
# Entrenamiento del modelo 1
history1 = model1.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=epochs,
    validation_data=val_generator,
    validation_steps=len(val_generator)
)
```

### Cargar los tres modelos:

```python
# Cargar modelos
model1 = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])
model1.load_weights('model1_weights.h5')

model2 = Sequential([
    Conv2D(64, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')
])
model2.load_weights('model2_weights.h5')

model3 = Sequential([
    Conv2D(128, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')
])
model3.load_weights('model3_weights.h5')
```

### PROCESO DE HIPERPARÁMETROS (TUNING)

```python
# Definir el data generators
batch_size = 32

train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_path,
    target_size=(150, 150),
    batch_size=batch_size,
    class_mode='binary'
)

val_generator = val_datagen.flow_from_directory(
    val_path,
    target_size=(150, 150),
    batch_size=batch_size,
    class_mode='binary'
)
```

### Definir la función de creación del modelo

```python
# Definir la función de creación del modelo
def create_model(hp):
    model = Sequential([
        Conv2D(hp.Int('conv1_units', min_value=32, max_value=128, step=32), (3, 3), activation='relu', input_shape=(150, 150, 3)),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer=Adam(learning_rate=hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    return model
```

### Verificar si la función construye el modelo correctamente

```python
create_model(keras_tuner.HyperParameters())
```

### Crear usar el tuner para afinar los hiperparámetros con RandomSearch

```python
# Configurar el sintonizador
tuner = keras_tuner.RandomSearch(
    hypermodel=create_model,
    objective="val_accuracy",
    max_trials=5,
    executions_per_trial=2,
    overwrite=True,
    directory='mi_direct3',
    project_name="helloworld3",
)
```

```python
# Imprimir en pantalla el espacio de búsqueda
tuner.search_space_summary()
```

### Buscar hiperparámetros

```python
# Buscar hiperparámetros
tuner.search(train_generator,
             epochs=10,
             validation_data=val_generator,
             steps_per_epoch=len(train_generator),
             validation_steps=len(val_generator))
```

```python
# Función para predecir utilizando los modelos individuales
def ensemble_predict(models, X):
    predictions = [model.predict(X) for model in models]
    return sum(predictions) / len(models)
```
